```python
from pymystem3 import Mystem

# сохраняем класс в переменную
mystem = Mystem() 
```

```python
with open ('test_my.txt', 'r+', encoding = 'utf-8') as file:
    text = file.read()
```

```python
from nltk import sent_tokenize
from nltk.tokenize import word_tokenize, wordpunct_tokenize
```

```python
import re
#normalized_text = lambda x: x in wordpunct_tokenize(text) if re.match('[!\”\"#$%&\'()»«*−–+—,-.·/:;<=>?@\\^_`{|}~?]',x) is None
#normalized_text
text = word_tokenize(text)
normalized_text = []
for i in text:
    if re.match('[!\”\"#$%&\'()»«*−–+—,-.·/:;<=>?@\\^_`{|}\[\]~?]', i) is None:
        string = str(i).lower()
        normalized_text.append(string)
#normalized_text
```

```python
normalized_text = ' '.join(normalized_text)
#normalized_text
```

```python
words_analized = mystem.analyze(normalized_text)
[words_analized.remove(i) for i in words_analized if i['text'] == ' ']#убираем словари с ключом "text", у которого значение пробел
words_analized
```

```python
with open ('my_stem_parse.txt', 'w', encoding = 'utf-8') as file:
       file.write(str(words_analized))
```

```python
for i in words_analized:
    #print(i)
    if i['text'].isalpha() and re.match('[а-я]+',i['text']) and i['analysis']!=[]:
        result = re.match('[A-Z]{1,6}', i['analysis'][0]['gr'])
        result = result.group(0)
        print('Токен - ', i['text'])
        print('Лемма слова - ',i['analysis'][0]['lex'])
        print('Часть речи - ',result)
        print('Грамматическое значение - ',i['analysis'][0]['gr'])
    #elif i['analysis']==[] and re.match('[Aa-Zz]+',i['text']):
        #print ('Иностранное слово -', i['text'])
    else:
        print('Другое - ', i)
```

```python
with open ('test_my_result.txt', 'w', encoding = 'utf-8') as file:
       file.write(str(words_analized))
```
