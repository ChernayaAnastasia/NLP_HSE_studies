```python
from nltk.util import ngrams
# ngrams - это встроенная функция в модуле nltk
from collections import Counter 
import re 
import pandas as pd 

df = pd.read_csv('DZ1811.csv', encoding="utf-8", sep=";")
df = df.dropna() #выкинет none значение из нашего датафрейма
#df
```

```python
df = df[df['description_locale'].notnull()]
#df #убираем нулевые значения (почему dropna не удалила?)
```

```python
df = df[df['description_locale'].str.match('en')] #нам нужны только строки, где значение en
#df
```

```python
titles = list(df["p3_summary_title"].values) #берем значения из колонки p3_summary_title
#titles
```

```python
preprocessed_titles = [str(i).lower().strip() for i in titles]
preprocessed_titles = [re.sub("[0123456789!\”\"#$%&\'()»«*−–+—,-.·/:;<=>?@[\\]^_`{|}~?⭐¦]","", i) for i in preprocessed_titles]
preprocessed_titles = [i.split() for i in preprocessed_titles]
new_list = []
for i  in preprocessed_titles:
    for word in i:
        #result =list(ngrams(i,3))
        new_list.append(word)        
```

```python
result =ngrams(new_list,3) #собираем 3-grams
result = Counter(result).most_common() #считаем самые частотные
sorted(result) #сортируем
result #показываем
```
